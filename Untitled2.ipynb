```python
import matplotlib.pyplot as plt

# Data
months = ['07/2019', '08/2019', '09/2019', '10/2019', '11/2019']
searches = [50, 53, 59, 56, 62]
direct = [39, 47, 42, 51, 51]
social_media = [70, 80, 90, 87, 92]

x = range(len(months))
bar_width = 0.25

# Plotting
plt.figure(figsize=(10, 6))
plt.bar([p - bar_width for p in x], searches, width=bar_width, label='Searches', color='dodgerblue')
plt.bar(x, direct, width=bar_width, label='Direct', color='lightcoral')
plt.bar([p + bar_width for p in x], social_media, width=bar_width, label='Social Media', color='orange')

plt.xticks(x, months)
plt.xlabel('Months')
plt.ylabel('Visitors in thousands')
plt.title('Visitors by web traffic sources')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
